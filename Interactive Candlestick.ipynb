# Gráfico de Candlestick Interativo

##### Entrada: Dados do Yahoo Finance | Saída: Gráfico de Candlestick Interativo

### 1. Bibliotecas

```python
# Importação de bibliotecas utilizadas
import plotly.graph_objects as go
import pandas as pd
from pandas_datareader import data
```

### 2. Obtenção de Dados

```python
# Criar variaveis com informações dos dados a serem obtidos
ativo = 'BTC-USD' # Ticker do ativo como no database
database = 'yahoo' # Fonte provedora de dados
start_date = '1/12/2020' # Data de início
end_date = '01/01/2021' # Data final

# Solicitação de dados
df = data.DataReader(ativo, database, start_date,end_date)

# Visualizar as 5 primeiras linhas 
df.head()
```

### 3. Plotagem de Gráfico Interativo

```python
# Criar gráfico com as informações obtidas
figure = go.Figure(
    data = [go.Candlestick(
                x = df.index, # Eixo X do gráfico (Série Temporal)
                low = df['Low'], # Preço Mínimo
                high = df['High'], # Preço Máximo
                close = df['Close'], # Preço de Fechamento
                open = df['Open'], # Preço de Abertura
                increasing_line_color = 'green',
                decreasing_line_color = 'red')]
)

# Exibição do gráfico interativo
figure.show()
```
